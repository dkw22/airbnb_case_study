```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('./AB_NYC_2019.csv')
air_bnb.head()
```

```python
# Which hosts are the busiest and why?
# The ones that have no availability.
```

```python
# How many neighborhood groups are available and which shows up the most?
air_bnb.groupby(['neighbourhood_group']).count()
# 5 are available and Manhattan shows up the most
```

```python
# Are private rooms the most popular in manhattan?
air_bnb[air_bnb['neighbourhood_group'] == 'Manhattan'].groupby('room_type').count()
```

```python
# Which hosts are the busiest and based on their reviews?
host_id_group = air_bnb.groupby(['host_id','host_name']).sum().sort_values('number_of_reviews',ascending=False)
host_id_group.head(10)
```

```python
#Which neighorhood group has the highest average price?
air_bnb.groupby(['neighbourhood_group']).mean()[['price']].sort_values('price',ascending=False)
```

```python
# Which neighbor hood group has the highest total price?
air_bnb.groupby(['neighbourhood_group']).sum()[['price']].sort_values('price',ascending=False)
```

```python
#Which top 5 hosts have the highest total price?
air_bnb.groupby(['host_id','host_name']).sum()[['price']].sort_values('price',ascending=False).head(5)
```

```python
air_bnb[(air_bnb['number_of_reviews'] >=100) & (air_bnb['availability_365'] == 0)].groupby('host_id').sum()
```

```python
# What host has the highest total of prices and where are they located?
air_bnb.groupby(['host_id','host_name','neighbourhood_group']).sum()[['price']].sort_values('price',ascending=False).head(1)
```

```python
# When did Danielle from Queens last receive a review?
air_bnb[air_bnb['host_name'] == 'Danielle'].sort_values('last_review',ascending=False)['last_review']
```
